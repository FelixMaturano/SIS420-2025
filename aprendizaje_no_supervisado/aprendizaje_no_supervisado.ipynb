```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Ruta local al dataset
folder_path = r"C:\Users\Felix\Downloads\all_images"

# Tamaño para redimensionar
image_size = (64, 64)

image_data = []
image_labels = []

label_dict = {}
label_id = 0

for file_name in os.listdir(folder_path):
    if file_name.lower().endswith((".jpg", ".jpeg", ".png")):
        file_path = os.path.join(folder_path, file_name)
        img = cv2.imread(file_path)
        img = cv2.resize(img, image_size)
        image_data.append(img)

        try:
            class_name = file_name.split("_")[-1].split(".")[0]
        except:
            class_name = "Unknown"

        if class_name not in label_dict:
            label_dict[class_name] = label_id
            label_id += 1

        image_labels.append(label_dict[class_name])
```

```python
# Mostrar cuántas imágenes se cargaron
print(f"Total de imágenes cargadas: {len(image_data)}")

# Mostrar las primeras 5 etiquetas (opcional)
print("Primeras 5 etiquetas (números de clase):", image_labels[:5])

# Mostrar un ejemplo de imagen cargada (opcional)
if len(image_data) > 0:
    plt.imshow(cv2.cvtColor(image_data[0], cv2.COLOR_BGR2RGB))
    plt.title("Ejemplo de imagen cargada")
    plt.axis('off')
    plt.show()
else:
    print("⚠️ No se cargaron imágenes. Verifica la ruta o el contenido.")
```
